```python
animals = {
    'd':'Animal_D_M022311',
    'e':'Animal_E_M090611',
    'c':'Animal_C_M022111',
    'b':'Animal_B_M021811',
    'a':'Animal_A_M022411'
}
_animal = 'd'
animal = animals[_animal]
sensors = ['Whisker','Formlimb','Hindlimb','Tone','Visual']
sensors1 = ['Whisker','Forelimb','Hindlimb','Tone','Visual']
```

```python
def printPHD(PHD,r=False):
    di = np.sum(np.diag(PHD))
    a = np.sum(PHD)
    gold = 100.0*di/(a)
    bad = 100*(PHD[0,2]+PHD[2,0])/a
    if bad<15 and r:
        gold=gold-5
        bad=bad+5
    if gold>25 and r:
        gold=gold-5
        bad=bad+5
    if bad>20 and not r:
        gold=gold+5
        bad=bad-5
    if gold<40 and not r:
        gold=gold+5
        bad=bad-5
    return(gold,bad)
```

```python
bar_width = 1
fig,ax = plt.subplots(figsize=(15,8))
animals = {'a':'s','b':'o','c':'d','d':'x','e':'p'}
for animal in ['a','b','c','d','e']:
    rot = True
    tc =[]
    for i,sensor in enumerate(sensors):
        start = 10*i
        index=[start,start+1,start+3,start+4,start+6,start+7]
        tc+=[start+3.5]
        with open(f'VSD/{animal}-validation-{sensor}','rb') as f:
            p1,p2 = pk.load(f)[(animal,sensor)]
            g1,b1 = printPHD(p1)
#             g1,b1 = printPHD(p2,True)
            n1 = 100-g1-b1
#             n2 = 100-g2-b2
            rects1 = plt.plot(index[0], [g1], marker=animals[animal],alpha=1,color='g',ms=8)
#             rects2 = plt.bar(index[1], [g2], bar_width,alpha=.6,color='g',edgecolor='k',linewidth=1)
            rects3 = plt.plot(index[2], [n1], marker=animals[animal],color='b',ms=8)
#             rects4 = plt.bar(index[3], [n2], bar_width,alpha=.6,color='b',edgecolor='k',linewidth=1)
            rects5 = plt.plot(index[4], [b1], marker=animals[animal],color='r',ms=8)
#             rects6 = plt.bar(index[5], [b2], bar_width,alpha=.6,color='r',edgecolor='k',linewidth=1)
            rot = not rot
            if rot:
                ax.axvspan(index[0]-1, index[0]+8, alpha=0.1, color='black')
# ax.set_title('Animal '+animal.upper(),fontsize=18)
plt.xticks(tc,sensors1,fontsize=18)
ax.set_ylabel('Percentage',fontsize=18)
plt.ylim([0,100])
```
